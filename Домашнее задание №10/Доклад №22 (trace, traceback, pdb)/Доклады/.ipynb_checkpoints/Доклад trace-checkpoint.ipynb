# Модуль trace  

https://pymotw.com/3/trace/index.html

Модуль трассировки полезен для понимания того, как выполняется программа. Он следит за выполнением операторов, создает отчеты  и помогает исследовать отношения между функциями, которые вызывают друг друга.

Эта программа будет использоваться в примерах, приведенных в остальной части рассказа.


```python
def recurse(level):
    print('recurse({})'.format(level))
    if level:
        recurse(level - 1)
```

Вывод трассировки:

```python
import trace
def main():
    print('This is the main program.')
    recurse(2)


if __name__ == '__main__':
    main()
```

trace.Trace - Создайте объект для отслеживания выполнения одного оператора или выражения. Все параметры являются необязательными. count обеспечивает подсчет числа линии. trace обеспечивает выполнение линии трассировки.

trace.run() - Выполняет команду и соберите статистику выполнения с текущими параметрами трассировки

```python
tracer = trace.Trace(count=False, trace=True)
tracer.run('recurse(2)')
```

runfunc() - тоже самое что run(), но исключительно для функций

```python
tracer = trace.Trace(count=False, trace=True)
tracer.runfunc(recurse, 2)
```

# Сохранение результатов

```python
tracer = trace.Trace(count=True, trace=False)
tracer.runfunc(recurse, 2)

results = tracer.results()
results.write_results(coverdir='ex1')
```

```python
tracer = trace.Trace(count=True,
                     trace=False,
                     outfile='trace_report.txt')
tracer.runfunc(recurse, 2)

report_tracer = trace.Trace(count=True,
                            trace=False,
                            infile='trace_report.txt')
results = tracer.results()
results.write_results(summary=True, coverdir='ex2')
```
