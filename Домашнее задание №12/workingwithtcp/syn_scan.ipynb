### Сначала разберёмся с датами и временем

```python
date_time = (pd.Period(datetime.datetime.now(), 'M') - 1).strftime("%Y-%m-%d %H:%M")    # пускай мы откатимся на недельну назад
formated_date_time = (pd.to_datetime(date_time) + pd.to_timedelta(df['time'].to_numpy(), 'D'))    # получили нужный для нас формат данных
```

```python
df['date'] = formated_date_time.strftime('%Y-%m-%d')    # нужный формат именно даты
df['time'] = formated_date_time.strftime('%H:%M:%S')    # нужный формат именно времени
df
```

```python
df = df.reindex(columns=['no.', 'date', 'time', 'source', 'destination', 'protocol', 'length', 'info', 'tcp_retransmission'])
df
```

### Работаем со столбцом _info_

Избавляемся от неприятного знака — '>'

```python
df['info'] = df['info'].str.replace(r" > ", "", regex=False)
```

#### Разберёмся с флагами, которые в скобочках

```python
def get_flags(row):
    return re.search(r"\[([A-Za-z0-9_, A-Za-z0-9_]+)\]", row).group(0)
```

```python
df['flags'] = df['info'].map(get_flags)
df
```

#### Теперь разберёмся с портами

```python
def get_port(row, port):
    return row.split()[:2][port] 
```

```python
df['source_port'] = df['info'].map(lambda r, p = 0: get_port(r, p))
df['destination_port'] = df['info'].map(lambda r, p = 1: get_port(r, p))
df
```

#### Посмотрим на атрибуты

Попробуем узнать сколько их всего

```python
attrs = [[attr.split('=')[0] for attr in row.split() if '=' in attr] for row in df['info'].to_list()]
```

```python
flatten = lambda l: [item for sublist in l for item in sublist]
```

```python
all_attrs = set(flatten(attrs))
all_attrs
```

```python
def get_attr(row, attr):
    return re.search(f'{attr}='+r'\d*', row).group(0).split('=')[-1] if attr in row else False
```

```python
for attr in ('Seq', 'Win', 'Len', 'MSS', 'Ack'):
    df[attr.lower()] = df['info'].map(lambda r, a=attr: get_attr(r, a))
df
```

```python
df = df.drop('info', 1)    # избавились он колонки info
df
```
